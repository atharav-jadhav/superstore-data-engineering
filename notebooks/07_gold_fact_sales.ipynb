# Fact table

```python
%sql
CREATE OR REPLACE TABLE superstoreprac.gold.fact_sales
AS
SELECT
    ROW_NUMBER() OVER (ORDER BY s.order_id, dp.product_key) AS sales_key,
    s.order_id ,
    dc.customer_key,
    dp.product_key,
    dl.location_key,
    od.date_key AS order_date_key,
    sd.date_key AS ship_date_key,
    s.sales,
    s.quantity,
    s.discount,
    s.profit,
    s.profit_margin,
    s.revenue_before_discount
FROM superstoreprac.silver.slv_table s
JOIN superstoreprac.gold.dim_customer dc
ON s.customer_id = dc.customer_id
JOIN  superstoreprac.gold.dim_product dp
ON s.product_id = dp.product_id
JOIN superstoreprac.gold.dim_location dl
ON s.city = dl.city 
AND s.state = dl.state
JOIN superstoreprac.gold.dim_date od
ON s.order_date = od.full_date
JOIN superstoreprac.gold.dim_date sd 
ON s.ship_date = sd.full_date
```

```python
%sql
SELECT COUNT(*) FROM superstoreprac.gold.fact_sales
```
