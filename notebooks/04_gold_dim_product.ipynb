```python
%sql
CREATE OR REPLACE TABLE superstoreprac.gold.dim_product
AS
WITH product AS (
    SELECT
        product_id,
        MAX(product_name)  AS product_name,
        MAX(category)      AS category,
        MAX(sub_category)  AS sub_category
    FROM superstoreprac.silver.slv_table
    GROUP BY product_id
)
SELECT
    ROW_NUMBER() OVER (ORDER BY product_id) AS product_key,
    product_id,
    product_name,
    category,
    sub_category
FROM product
```
