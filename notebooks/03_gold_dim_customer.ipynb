## dim customers

```python
%sql

CREATE OR REPLACE TABLE superstoreprac.gold.dim_customer
AS
WITH customers AS (
    SELECT DISTINCT
        customer_id,
        customer_name,
        segment
    FROM superstoreprac.silver.slv_table
)

SELECT
    ROW_NUMBER() OVER (ORDER BY customer_id) AS customer_key,
    customer_id,
    customer_name,
    segment
FROM customers;
```

```python
%sql
SELECT COUNT(*) FROM superstoreprac.gold.dim_customer
```
