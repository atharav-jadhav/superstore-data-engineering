# dim location

```python
%sql
CREATE OR REPLACE TABLE superstoreprac.gold.dim_location
AS
WITH location AS (
    SELECT DISTINCT
        city,
        state,
        region,
        country
    FROM superstoreprac.silver.slv_table
)
SELECT
    ROW_NUMBER() OVER (ORDER BY state, city) AS location_key,
    city,
    state,
    region,
    country
FROM location
```

```python
%sql
SELECT COUNT(*) FROM superstoreprac.gold.dim_location
```
